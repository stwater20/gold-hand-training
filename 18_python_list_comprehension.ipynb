## 1. 建立 List


````python
a = [1, 2, 3, 4, 5]          # 直接建立
b = list(range(5))            # 從 range 建立
c = [0] * 5                   # 5 個 0
d = list("hello")             # 字串拆成字元
print(a, b, c, d, sep="\n")
````

## 2. 讀取 List 中的元素


````python
a = [10, 20, 30, 40, 50]
print(a[0])      # 第一個
print(a[-1])     # 最後一個
print(a[1:4])    # 切片：index 1 到 3
print(a[:3])     # 前三個
print(a[::2])    # 每隔一個取
````

## 3. 讀取 List 的逆序元素


````python
a = [10, 20, 30, 40, 50]
print(a[::-1])            # 切片逆序（產生新 list）
print(list(reversed(a)))  # reversed（iterator）

for x in reversed(a):
    print(x, end=" ")
print()
print(a[-2])              # 倒數第二個
````

## 4. List Comprehension

用一行產生 list，比迴圈 `append` 快也更好讀。


````python
# 二維矩陣
matrix = [[i + j for j in range(3)] for i in range(3)]
print(matrix)
# 過程：
# i = 0: [0+0, 0+1, 0+2] = [0, 1, 2]
# i = 1: [1+0, 1+1, 1+2] = [1, 2, 3]
# i = 2: [2+0, 2+1, 2+2] = [2, 3, 4]
````

````python
# 條件過濾
evens = [x for x in range(10) if x % 2 == 0]
print(evens)
````

````python
# 雙重條件：連續兩個 if 等於 and
result = [x for x in range(100) if x % 2 == 0 if x % 3 == 0]
result2 = [x for x in range(100) if x % 2 == 0 and x % 3 == 0]
print(result)
print(result == result2)
````

````python
# 攤平二維 list
matrix = [[i + j for j in range(3)] for i in range(3)]
flat = [item for sublist in matrix for item in sublist]

# 等價寫法
flat2 = []
for sublist in matrix:
    for item in sublist:
        flat2.append(item)

print(flat)
print(flat == flat2)
````

## 6. 建立多維的 list

**注意陷阱**：`[[0] * C] * R` 是 R 個「同一列」的參考！
（詳見 `0_python_basics.ipynb` 第 7 節）


````python
R, C = 3, 4

# 錯誤寫法：三列是同一個物件
bad = [[0] * C] * R
bad[0][0] = 99
print("錯誤寫法:", bad)

# 正確寫法
good = [[0] * C for _ in range(R)]
good[0][0] = 99
print("正確寫法:", good)
````

## 7. 讀取多維 List 中的內容


````python
grid = [[1, 2, 3],
        [4, 5, 6],
        [7, 8, 9]]

print(grid[1][2])         # 第 1 列第 2 行 → 6

for row in grid:          # 逐列走訪
    print(row)

for i in range(len(grid)):        # 用索引走訪
    for j in range(len(grid[0])):
        print(grid[i][j], end=" ")
    print()
````

## 8. 應用：迷宮走向問題（簡化版）

【問題描述】給出一個迷宮，每個位置上有一個人，每個位置有指示牌：
`L` 向左、`R` 向右、`U` 向上、`D` 向下。
走出邊界代表「走出迷宮」；如果走回走過的位置代表繞圈，永遠出不來。
請問有多少人能走出來？

【範例輸入】

```text
3 3
RRD
URD
ULL
```


````python
sample = "3 3\nRRD\nURD\nULL"
data = sample.split("\n")

R, C = map(int, data[0].split())
grid = [data[i + 1] for i in range(R)]

move = {'L': (0, -1), 'R': (0, 1), 'U': (-1, 0), 'D': (1, 0)}

def can_escape(r, c):
    visited = set()
    while 0 <= r < R and 0 <= c < C:
        if (r, c) in visited:      # 繞圈了
            return False
        visited.add((r, c))
        dr, dc = move[grid[r][c]]
        r, c = r + dr, c + dc
    return True

ans = sum(can_escape(i, j) for i in range(R) for j in range(C))
print(f"{ans} 人可以走出迷宮（共 {R * C} 人）")
````
