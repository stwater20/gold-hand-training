## 1. Python 版本：大量迴圈計時

先用 Python 跑一個簡單的累加迴圈。

- Python 逐步執行迴圈的成本比較高
- 大量迴圈時，Python 通常會比 C / C++ 慢
- 程式競賽使用 Python 時，要更注意時間複雜度

````python
from time import time

k = 0
n = 10_000_000

start = time()

for i in range(n):
    k += 1

end = time()

print("k =", k)
print("Python 執行時間:", end - start, "秒")
````

## 2.  C 程式檔案

`%%writefile speed.c` 的意思是：

> 把這個 cell 裡面的內容寫入 `speed.c` 檔案

這個 cell 不會直接執行 C 程式，只是產生一個 C 原始碼檔。

````python
%%writefile speed.c
#include <stdio.h>
#include <time.h>

int main() {
    int k = 0;
    int n = 10000000;

    clock_t start, end;
    start = clock();

    for (int i = 0; i < n; i++) {
        k++;
    }

    end = clock();

    printf("k = %d\n", k);
    printf("C 執行時間: %.6f 秒\n", (double)(end - start) / CLOCKS_PER_SEC);

    return 0;
}
````

## 3. 編譯並執行 C 程式

產生 `speed.c` 後，要先編譯：

```bash
gcc speed.c -o speed_c
```

把 `speed.c` 編譯成可執行檔 `speed_c`。

接著執行：

```bash
./speed_c
```

````python
!gcc speed.c -o speed_c
!./speed_c
````

## 4. 在 Jupyter 中寫出 C++ 程式檔案

同樣使用 `%%writefile`，只是這次寫成 `speed.cpp`。

C++ 常見於程式設計競賽，速度通常比 Python 快很多。

````python
%%writefile speed.cpp
#include <bits/stdc++.h>
using namespace std;

int main() {
    int k = 0;
    int n = 10000000;

    clock_t start, end;
    start = clock();

    for (int i = 0; i < n; i++) {
        k++;
    }

    end = clock();

    cout << "k = " << k << "\n";
    cout << "C++ 執行時間: " << fixed << setprecision(6)
         << (double)(end - start) / CLOCKS_PER_SEC << " 秒\n";

    return 0;
}
````

## 5. 編譯並執行 C++ 程式

```bash
g++ speed.cpp -o speed_cpp
./speed_cpp
```

在不同電腦上，時間數字可能不一樣；重點是觀察 Python 與 C++ 的差距。

````python
!g++ speed.cpp -o speed_cpp
!./speed_cpp
````
