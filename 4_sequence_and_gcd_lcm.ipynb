# Part 1：等差數列

## 什麼是等差數列？

如果一個數列中，任意兩個相鄰項目的「差」都相同，這個數列就是**等差數列**。

這個固定的差稱為**公差**，通常使用 $d$ 表示。

例如：

$$2, 5, 8, 11, 14$$

檢查相鄰兩項：

$$5-2=3$$
$$8-5=3$$
$$11-8=3$$
$$14-11=3$$

每次都增加 3，所以這是等差數列，公差為：

$$d=3$$


```python
# 使用 Python 觀察等差數列的相鄰差值
arithmetic_sequence = [2, 5, 8, 11, 14]

for index in range(len(arithmetic_sequence) - 1):
    difference = arithmetic_sequence[index + 1] - arithmetic_sequence[index]
    print(
        f"{arithmetic_sequence[index + 1]} - "
        f"{arithmetic_sequence[index]} = {difference}"
    )
```

## 公差也可以是負數或零

公差為負數：

$$10, 7, 4, 1, -2$$

每次都減少 3，因此公差是：

$$d=-3$$

公差為零：

$$6, 6, 6, 6$$

每次的差都是 0，因此公差是：

$$d=0$$

## 等差數列第 n 項公式

如果第一項是 $a_1$，公差是 $d$，第 $n$ 項為：

$$a_n=a_1+(n-1)d$$

例如第一項是 2、公差是 3，第 5 項為：

$$a_5=2+(5-1)\times3=14$$


```python
def arithmetic_nth(first: int, difference: int, n: int) -> int:
    '''計算等差數列的第 n 項。'''
    return first + (n - 1) * difference


print(arithmetic_nth(first=2, difference=3, n=5))
assert arithmetic_nth(2, 3, 5) == 14
assert arithmetic_nth(10, -3, 4) == 1
```

# Part 2：等比數列

## 什麼是等比數列？

如果一個數列中，後一項除以前一項所得的「比值」都相同，這個數列就是**等比數列**。

這個固定的比值稱為**公比**，通常使用 $r$ 表示。

例如：

$$2, 6, 18, 54, 162$$

檢查相鄰兩項：

$$6\div2=3$$
$$18\div6=3$$
$$54\div18=3$$
$$162\div54=3$$

每一項都是前一項的 3 倍，所以這是等比數列，公比為：

$$r=3$$


```python
# 使用 Python 觀察等比數列的相鄰比值
geometric_sequence = [2, 6, 18, 54, 162]

for index in range(len(geometric_sequence) - 1):
    ratio = geometric_sequence[index + 1] / geometric_sequence[index]
    print(
        f"{geometric_sequence[index + 1]} / "
        f"{geometric_sequence[index]} = {ratio:g}"
    )
```

## 公比也可以是小數、負數或零

公比為小數：

$$16, 8, 4, 2, 1$$

每一項都是前一項的 $\frac{1}{2}$，所以公比是：

$$r=\frac{1}{2}$$

公比為負數：

$$2, -4, 8, -16$$

正負號交替，每一項都是前一項的 $-2$ 倍：

$$r=-2$$

公比為零：

$$5, 0, 0, 0$$

第一項乘以 0 後，後面的項目都會是 0。

## 等比數列第 n 項公式

如果第一項是 $a_1$，公比是 $r$，第 $n$ 項為：

$$a_n=a_1\times r^{n-1}$$

例如第一項是 2、公比是 3，第 5 項為：

$$a_5=2\times3^{4}=162$$


```python
def geometric_nth(first: float, ratio: float, n: int) -> float:
    '''計算等比數列的第 n 項。'''
    return first * ratio ** (n - 1)


print(geometric_nth(first=2, ratio=3, n=5))
assert geometric_nth(2, 3, 5) == 162
assert geometric_nth(2, -2, 4) == -16
```

## 等差與等比怎麼分辨？

| 比較項目 | 等差數列 | 等比數列 |
|---|---|---|
| 觀察方式 | 相鄰兩項相減 | 後一項除以前一項 |
| 固定的數 | 公差 $d$ | 公比 $r$ |
| 第 n 項 | $a_1+(n-1)d$ | $a_1r^{n-1}$ |
| 成長方式 | 每次增加或減少固定數量 | 每次乘上固定倍數 |

例子：

- `4, 7, 10, 13`：差值都是 3，所以是等差數列。
- `4, 8, 16, 32`：比值都是 2，所以是等比數列。
- `1, 2, 4, 7`：差值與比值都不固定，所以兩者都不是。

> 常數數列如 `5, 5, 5, 5` 的公差是 0、公比是 1，因此它可以同時被視為等差與等比數列。

# Part 3：最大公因數 GCD

## 先認識因數

如果整數 $a$ 可以整除整數 $b$，那麼 $a$ 就是 $b$ 的**因數**。

例如 12 的正因數有：

$$1, 2, 3, 4, 6, 12$$

18 的正因數有：

$$1, 2, 3, 6, 9, 18$$

12 和 18 共同擁有的因數是：

$$1, 2, 3, 6$$

其中最大的是 6，因此：

$$\operatorname{GCD}(12,18)=6$$

GCD 是 Greatest Common Divisor 的縮寫，中文是**最大公因數**。


```python
def positive_factors(number: int) -> list[int]:
    '''列出正整數的所有正因數。'''
    return [value for value in range(1, number + 1) if number % value == 0]


factors_of_12 = positive_factors(12)
factors_of_18 = positive_factors(18)
common_factors = sorted(set(factors_of_12) & set(factors_of_18))

print("12 的因數：", factors_of_12)
print("18 的因數：", factors_of_18)
print("公因數：", common_factors)
print("最大公因數：", max(common_factors))
```

## GCD 的生活例子

有 12 顆蘋果與 18 顆橘子，要平均分成數量最多的相同禮物袋，而且每袋的蘋果數與橘子數都相同。

最多可以分成：

$$\operatorname{GCD}(12,18)=6\text{ 袋}$$

每袋有：

$$12\div6=2\text{ 顆蘋果}$$
$$18\div6=3\text{ 顆橘子}$$

## 特別情況

- `GCD(a,0) = |a|`，例如 `GCD(8,0)=8`。
- GCD 通常使用非負數表示。
- `GCD(0,0)` 在不同教材中可能有不同約定，使用前應先確認題目定義。


```python
import math

print("GCD(12,18) =", math.gcd(12, 18))
print("GCD(8,0) =", math.gcd(8, 0))
print("GCD(-12,18) =", math.gcd(-12, 18))

assert math.gcd(12, 18) == 6
assert math.gcd(8, 0) == 8
```

# Part 4：最小公倍數 LCM

## 先認識倍數

一個整數乘以 1、2、3、……得到的數，稱為這個整數的**倍數**。

12 的正倍數：

$$12, 24, 36, 48, 60,ldots$$

18 的正倍數：

$$18, 36, 54, 72,ldots$$

12 和 18 第一個共同出現的正倍數是 36，因此：

$$\operatorname{LCM}(12,18)=36$$

LCM 是 Least Common Multiple 的縮寫，中文是**最小公倍數**。


```python
multiples_of_12 = [12 * value for value in range(1, 7)]
multiples_of_18 = [18 * value for value in range(1, 7)]

print("12 的前幾個倍數：", multiples_of_12)
print("18 的前幾個倍數：", multiples_of_18)

common_multiples = sorted(set(multiples_of_12) & set(multiples_of_18))
print("目前找到的公倍數：", common_multiples)
print("最小公倍數：", min(common_multiples))
```

## LCM 的生活例子

一盞燈每 12 秒閃一次，另一盞燈每 18 秒閃一次。假設現在兩盞燈同時閃，下一次同時閃是在：

$$\operatorname{LCM}(12,18)=36\text{ 秒後}$$

## GCD 與 LCM 的關係

對兩個非零整數 $a$、$b$：

$$\operatorname{GCD}(a,b)\times\operatorname{LCM}(a,b)=|a\times b|$$

因此：

$$\operatorname{LCM}(a,b)=\frac{|a\times b|}{\operatorname{GCD}(a,b)}$$

以 12 和 18 為例：

$$6\times36=216$$
$$12\times18=216$$

兩邊相等。


```python
gcd_value = math.gcd(12, 18)
lcm_value = math.lcm(12, 18)

print("GCD(12,18) =", gcd_value)
print("LCM(12,18) =", lcm_value)
print("GCD × LCM =", gcd_value * lcm_value)
print("12 × 18 =", 12 * 18)

assert gcd_value * lcm_value == abs(12 * 18)
```

## GCD 與 LCM 怎麼分辨？

| 比較項目 | GCD | LCM |
|---|---|---|
| 中文 | 最大公因數 | 最小公倍數 |
| 尋找對象 | 共同的因數 | 共同的倍數 |
| 選擇方式 | 選最大的 | 選最小的正數 |
| 常見用途 | 平均分組、切割成最大相同單位 | 找共同週期、同時發生的最早時間 |

記憶方式：

- GCD：要「分東西」，找最大的相同分組單位。
- LCM：要「等週期重逢」，找最早共同出現的時間。

# 總結

- **等差數列**：相鄰兩項的差固定。
- **等比數列**：相鄰兩項的比值固定。
- **GCD**：兩個數共同因數中最大的。
- **LCM**：兩個數共同倍數中最小的正數。

看到數列時，先問「固定的是差，還是比？」

看到兩個整數時，先問「要平均分組，還是找共同週期？」
