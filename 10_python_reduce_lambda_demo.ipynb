# reduce 及 lambda 教學示範

- `lambda`：匿名函式
- `reduce()`：把一串資料「累積」成一個結果
- 常見應用：總和、乘積、最大值、GCD、all / any、Fibonacci


## 1. 先認識 lambda

`lambda` 是一種簡短函式寫法。

一般函式：

```python
def add(x, y):
    return x + y
```

可以寫成：

```python
lambda x, y: x + y
```

格式是：

```python
lambda 參數1, 參數2: 回傳值
```

````python
def add(x, y):
    return x + y

add_lambda = lambda x, y: x + y

print(add(3, 5))
print(add_lambda(3, 5))
````

## 2. reduce 是什麼？

`reduce()` 會把序列中的元素，**從左到右兩兩合併**，最後變成一個結果。

例如：

```python
reduce(lambda x, y: x + y, [1, 2, 3, 4, 5])
```

過程可以想成：

```text
先算 1 + 2 = 3
再算 3 + 3 = 6
再算 6 + 4 = 10
再算 10 + 5 = 15
```

所以最後結果是 `15`。

````python
from functools import reduce

result = reduce(lambda x, y: x + y, [1, 2, 3, 4, 5])
print(result)
````

## 3. 把 reduce 過程印出來

下面用一般函式取代 `lambda`，讓學生看清楚 `x` 和 `y` 分別是什麼。

`x`：目前累積的結果  
`y`：下一個要處理的元素

````python
from functools import reduce

def show_add(x, y):
    print(f"x={x}, y={y}, x+y={x+y}")
    return x + y

result = reduce(show_add, [1, 2, 3, 4, 5])
print("最後結果:", result)
````

## 4. 計算序列中所有元素的乘積

用 `reduce()` 做乘積：

```python
product = reduce(lambda x, y: x * y, numbers)
```

過程：

```text
1 * 2 = 2
2 * 3 = 6
6 * 4 = 24
24 * 5 = 120
```

````python
from functools import reduce

numbers = [1, 2, 3, 4, 5]
product = reduce(lambda x, y: x * y, numbers)
print("乘積:", product)
````

## 5. 找出序列中最大值

邏輯：

```python
x if x > y else y
```

意思是：

> 如果 `x` 比 `y` 大，就留下 `x`，否則留下 `y`。

````python
from functools import reduce

numbers = [1, 9, 3, 7, 5]
max_value = reduce(lambda x, y: x if x > y else y, numbers)
print("最大值:", max_value)

# 實務上更建議直接用 max()
print("用 max():", max(numbers))
````

## 6. 用 reduce 計算多個數字的最大公因數 GCD

兩個數的 GCD 可以用：

```python
math.gcd(a, b)
```

多個數字的 GCD 可以用 `reduce()` 把它們一路合併。

````python
from functools import reduce
import math

nums = [24, 36, 60]
gcd_value = reduce(math.gcd, nums)
print("最大公因數:", gcd_value)
````

## 7. 判斷序列中元素是否全部為正數

這裡用 `reduce()` 模擬 `all()`：

```python
all_positive = reduce(lambda x, y: x and (y > 0), nums, True)
```

注意第三個參數 `True` 是初始值。

流程：

```text
一開始 x = True
檢查第一個數字是否 > 0
再檢查第二個數字是否 > 0
一路檢查到最後
```

````python
from functools import reduce

nums = [1, 2, 3, 4, 5]
all_positive = reduce(lambda x, y: x and (y > 0), nums, True)
print("全部為正數:", all_positive)

# 實務上更建議直接用 all()
print("用 all():", all(x > 0 for x in nums))
````

## 8. 判斷序列中是否存在任一偶數

這裡用 `reduce()` 模擬 `any()`：

```python
any_even = reduce(lambda x, y: x or (y % 2 == 0), nums, False)
```

第三個參數 `False` 是初始值。

````python
from functools import reduce

nums = [1, 3, 5, 7, 8]
any_even = reduce(lambda x, y: x or (y % 2 == 0), nums, False)
print("存在偶數:", any_even)

# 實務上更建議直接用 any()
print("用 any():", any(x % 2 == 0 for x in nums))
````

## 9. reduce 的初始值 initial

`reduce()` 可以放第三個參數，代表「一開始的累積值」。

```python
reduce(function, iterable, initial)
```

例如：

```python
reduce(lambda x, y: x + y, [1, 2, 3], 100)
```

過程：

```text
100 + 1 = 101
101 + 2 = 103
103 + 3 = 106
```

````python
from functools import reduce

result = reduce(lambda x, y: x + y, [1, 2, 3], 100)
print(result)
````

## 10. 空 list 的問題

如果沒有初始值，`reduce()` 遇到空 list 會錯。

````python
from functools import reduce

try:
    print(reduce(lambda x, y: x + y, []))
except TypeError as e:
    print("錯誤:", e)

# 有 initial 就不會錯
print(reduce(lambda x, y: x + y, [], 0))
````

## 11. 用 reduce 產生 Fibonacci 數列


核心想法：

```python
x + [x[-1] + x[-2]]
```

如果目前是：

```python
[0, 1]
```

下一步會變成：

```python
[0, 1, 1]
```

再下一步：

```python
[0, 1, 1, 2]
```

````python
from functools import reduce

def fibonacci_list(n):
    if n <= 0:
        return []
    if n == 1:
        return [0]
    return reduce(lambda x, _: x + [x[-1] + x[-2]], range(n - 2), [0, 1])

print(fibonacci_list(10))
````

## 12. 如果只要第 n 個 Fibonacci

上面那種每次 `x + [...]` 都會建立新 list，資料大時不划算。

比較直覺、比賽也比較好寫的版本：

````python
def fibonacci(n):
    if n == 0:
        return 0
    a, b = 0, 1
    for _ in range(n - 1):
        a, b = b, a + b
    return b

for i in range(10):
    print(i, fibonacci(i))
````

## 13. reduce vs 內建函式

很多情況下，Python 已經有更清楚的內建函式。

| 目的 | reduce 寫法 | 建議寫法 |
|---|---|---|
| 加總 | `reduce(lambda x,y: x+y, nums)` | `sum(nums)` |
| 最大值 | `reduce(lambda x,y: x if x>y else y, nums)` | `max(nums)` |
| 最小值 | `reduce(lambda x,y: x if x<y else y, nums)` | `min(nums)` |
| 全部符合 | `reduce(lambda x,y: x and 條件, nums, True)` | `all(...)` |
| 任一符合 | `reduce(lambda x,y: x or 條件, nums, False)` | `any(...)` |
| 多數字 GCD | `reduce(math.gcd, nums)` | `reduce(math.gcd, nums)` 很適合 |

## 14. 小型速度比較：reduce vs sum

這裡讓學生看到：

> 會寫 `reduce()` 不代表它一定是最快或最適合。

````python
from functools import reduce
from time import perf_counter

nums = list(range(1_000_000))

start = perf_counter()
ans1 = reduce(lambda x, y: x + y, nums)
end = perf_counter()
print("reduce:", ans1, f"{end - start:.6f} 秒")

start = perf_counter()
ans2 = sum(nums)
end = perf_counter()
print("sum   :", ans2, f"{end - start:.6f} 秒")
````

## 15. 課堂練習 1：用 reduce 算平方和

給定：

```python
nums = [1, 2, 3, 4]
```

請算：

```text
1^2 + 2^2 + 3^2 + 4^2 = 30
```

提示：可以使用 initial。

````python
from functools import reduce

nums = [1, 2, 3, 4]
answer = reduce(lambda acc, x: acc + x ** 2, nums, 0)
print(answer)
````

## 16. 課堂練習 2：用 reduce 找最長字串

給定：

```python
words = ["cat", "elephant", "dog", "tiger"]
```

請找出最長的字串。

````python
from functools import reduce

words = ["cat", "elephant", "dog", "tiger"]
longest = reduce(lambda x, y: x if len(x) >= len(y) else y, words)
print(longest)
````

## 17. 課堂練習 3：用 reduce 計算所有數字的 LCM

LCM 是最小公倍數。

公式：

```python
lcm(a, b) = abs(a * b) // gcd(a, b)
```

給定：

```python
nums = [4, 6, 8]
```

答案應該是 `24`。

````python
from functools import reduce
import math

def lcm(a, b):
    return abs(a * b) // math.gcd(a, b)

nums = [4, 6, 8]
answer = reduce(lcm, nums)
print(answer)
````

## 18. 重點

1. `lambda x, y: x + y` 是簡短函式。
2. `reduce()` 是把很多資料累積成一個結果。
3. `reduce()` 的 `x` 通常是目前累積值，`y` 是下一個元素。
4. 能用 `sum()`、`max()`、`min()`、`all()`、`any()` 時，通常優先用它們。
5. `reduce()` 很適合用在「兩兩合併」的問題，例如多個數字的 GCD / LCM。
