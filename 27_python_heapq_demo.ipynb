`heapq` 是「最小堆積」：`heappop` 永遠彈出目前最小的元素，
插入與彈出都是 O($\log n$)，比每次都 `sort()` 快得多。


## 1. 基本操作


````python
import heapq

li = [5, 7, 9, 1, 3]
heapq.heapify(li)               # 就地轉成堆積
print("Heap : ", end="")
print(list(li))

heapq.heappush(li, 4)           # 插入
print("After push : ", end="")
print(list(li))

print("Pop 最小值 : ", heapq.heappop(li))   # 彈出最小
print("剩下 : ", list(li))
````

注意：堆積印出來「不是完全排序」的，
它只保證 `li[0]` 是最小值（父節點 ≤ 子節點）。


## 2. `nlargest` 與 `nsmallest`


````python
import heapq

grades = [110, 25, 38, 49, 20, 95, 33, 87, 80, 90]
print(heapq.nlargest(3, grades))    # 前 3 大
print(heapq.nsmallest(4, grades))   # 前 4 小
````

## 3. 想要「最大堆積」怎麼辦？

`heapq` 只有最小堆積。技巧：**存負數**。


````python
import heapq

nums = [5, 7, 9, 1, 3]
max_heap = [-x for x in nums]
heapq.heapify(max_heap)

print(-heapq.heappop(max_heap))   # 最大值 9
print(-heapq.heappop(max_heap))   # 次大值 7
````

## 4. HackerRank - Jesse and Cookies

題目連結：<https://www.hackerrank.com/challenges/jesse-and-cookies/problem>

【題目描述】每個曲奇都有甜度值。為了使所有曲奇甜度都至少 $K$，可重複操作：
取出兩個甜度最低的曲奇 $A$、$B$，做成新曲奇，甜度 = $A + 2B$（$A \le B$）。
求最少操作次數；若無法達成輸出 -1。

【範例輸入】

```text
6 7
1 2 3 9 10 12
```

【範例輸出】

```text
2
```


````python
import heapq

sample = "6 7\n1 2 3 9 10 12"
data = sample.split("\n")

def cookies(k, A):
    heapq.heapify(A)
    ops = 0
    while A[0] < k:
        if len(A) < 2:          # 剩不到兩個還達不到 → 失敗
            return -1
        a = heapq.heappop(A)    # 最小
        b = heapq.heappop(A)    # 次小
        heapq.heappush(A, a + 2 * b)
        ops += 1
    return ops

n_, k = map(int, data[0].split())
A = list(map(int, data[1].split()))
print(cookies(k, A))    # 應輸出 2
````

## 5. 為什麼用 heapq 而不是每次 sort？

每次操作都要拿「最小的兩個」：

- 每次 `sort()`：一次 O($n \log n$)，操作 $n$ 次 → O($n^2 \log n$)
- `heapq`：一次 O($\log n$)，操作 $n$ 次 → O($n \log n$)


````python
import heapq, random
from time import perf_counter

n_ = 20000
data_ = [random.randint(1, 10**6) for _ in range(n_)]

# heapq 版
A = data_[:]
start = perf_counter()
heapq.heapify(A)
while len(A) > 1:
    a = heapq.heappop(A)
    b = heapq.heappop(A)
    heapq.heappush(A, a + b)
print(f"heapq 版: {perf_counter() - start:.4f} 秒")

# 每次 sort 版
A = data_[:]
start = perf_counter()
while len(A) > 1:
    A.sort(reverse=True)
    a = A.pop()
    b = A.pop()
    A.append(a + b)
print(f"sort  版: {perf_counter() - start:.4f} 秒")
````

## 6. `heapq` 的常見應用

- **處理優先級佇列問題**：需要根據優先級處理任務時
- **Dijkstra 演算法**：圖論最短路徑，用 heap 挑「目前距離最小」的節點
- **合併排序列表**：`heapq.merge()` 可以合併多個已排序的列表
- **Top-K 問題**：`nlargest` / `nsmallest`


````python
import heapq

# heapq.merge：合併多個已排序的列表（lazy，省記憶體）
a = [1, 4, 7]
b = [2, 5, 8]
c = [3, 6, 9]
print(list(heapq.merge(a, b, c)))
````

## 7. 練習題

【LeetCode Task Scheduler 簡化版】給一串任務與冷卻時間 $n$，
相同任務之間至少間隔 $n$ 個時間單位，求完成所有任務的最短時間。

輸入：`tasks = ["A","A","A","B","B","B"]`, `n = 2`

期望輸出：

```text
8
```

（A → B → idle → A → B → idle → A → B）

提示：數出現最多次的任務次數 `max_c` 與有幾種任務同為 `max_c` 次，
答案 = `max(len(tasks), (max_c - 1) * (n + 1) + 同為最多次的任務種數)`。


````python
# 練習作答區
tasks = ["A", "A", "A", "B", "B", "B"]
n_cool = 2
````

### 參考答案

````python
from collections import Counter

tasks = ["A", "A", "A", "B", "B", "B"]
n_cool = 2

c = Counter(tasks)
max_c = max(c.values())
same = sum(1 for v in c.values() if v == max_c)
print(max(len(tasks), (max_c - 1) * (n_cool + 1) + same))
````
