## 1. `math` 常用函式


````python
import math

print(math.gcd(2048, 3072))   # 最大公因數 → 1024
print(math.perm(4, 2))        # P(4,2) 排列數 → 12
print(math.comb(5, 3))        # C(5,3) 組合數 → 10
print(math.lcm(6, 9))         # 最小公倍數 → 18（Python 3.9+）
print(math.isqrt(17))         # 整數平方根 → 4（不會有浮點誤差）
````

## 2. GCD 與 LCM

輾轉相除法：$\gcd(a, b) = \gcd(b, a \bmod b)$

$\text{lcm}(a, b) = \dfrac{a \times b}{\gcd(a, b)}$


````python
# 自己寫輾轉相除法（理解原理用，比賽直接用 math.gcd）
def my_gcd(a, b):
    while b:
        a, b = b, a % b
    return a

print(my_gcd(48, 36))
print(my_gcd(6, 11))

def my_lcm(a, b):
    return a * b // my_gcd(a, b)

print(my_lcm(6, 9))
````

````python
# 多個數字的 GCD：用 reduce 串起來
import math
from functools import reduce

nums = [24, 36, 60]
gcd_value = reduce(math.gcd, nums)
print("最大公因數:", gcd_value)

lcm_value = reduce(math.lcm, nums)
print("最小公倍數:", lcm_value)
````

## 3. 質數判斷：試除法

判斷單一數字 $n$ 是否為質數：檢查 $2 \sim \sqrt{n}$ 是否整除，O($\sqrt{n}$)。


````python
def is_prime(n_):
    if n_ < 2:
        return False
    for i in range(2, int(n_ ** 0.5) + 1):
        if n_ % i == 0:
            return False
    return True

print([x for x in range(2, 30) if is_prime(x)])
print(is_prime(65537))
````

## 4. 埃拉托斯特尼篩法建表

要判斷「很多個」數是不是質數時，先一次建好質數表：

1. 先假設 2 以上全部是質數
2. 從 2 開始，把每個質數的倍數全部劃掉
3. 沒被劃掉的就是質數

時間複雜度 O($n \log \log n$)，比一個一個試除快得多。


````python
u = 2 ** 15 + 1
prime = [True] * u
prime[0] = prime[1] = False
for p in range(2, u):
    if prime[p]:
        s = 2 * p
        for i in range(s, u, p):    # p 的倍數都不是質數
            prime[i] = False

print("2^15 以內的質數個數:", sum(prime))
print("前 10 個質數:", [i for i in range(u) if prime[i]][:10])
print("32749 是質數嗎?", prime[32749])
````

## 5. 建表 vs 逐一判斷的速度比較


````python
from time import perf_counter

u = 100000

# 方法一：逐一試除
start = perf_counter()
cnt1 = sum(1 for x in range(u) if is_prime(x))
t1 = perf_counter() - start

# 方法二：篩法建表
start = perf_counter()
prime = [True] * u
prime[0] = prime[1] = False
for p in range(2, u):
    if prime[p]:
        for i in range(2 * p, u, p):
            prime[i] = False
cnt2 = sum(prime)
t2 = perf_counter() - start

print(f"逐一試除: {cnt1} 個, {t1:.4f} 秒")
print(f"篩法建表: {cnt2} 個, {t2:.4f} 秒")
````

## 6. 實戰：GCD 與 LCM 題

【題目】給定兩個整數 $a$、$b$，求 GCD 與 LCM。
若 $a = b = 0$，輸出 `no GCD` / `no LCM`。
注意 $\gcd(0, x) = x$，而任何數與 0 的 LCM 是 0。

【範例輸入】

```text
6 11
0 0
6 9
```


````python
import math

sample = "6 11\n0 0\n6 9"
data = sample.split("\n")

for idx, line in enumerate(data, 1):
    a, b = map(int, line.split())
    print(f"Case {idx}:")
    if a == 0 and b == 0:
        print("no GCD")
        print("no LCM")
    else:
        print(f"GCD({a},{b}) = {math.gcd(a, b)}")
        print(f"LCM({a},{b}) = {math.lcm(a, b)}")
````
