## 1. `map()` 是什麼？

`map(function, iterable)` 的意思是：

> 把一個函式套用到每一個元素上。

例如：

```python
map(int, ['1', '2', '3'])
```

意思是：

```python
int('1'), int('2'), int('3')
```

最後會得到：

```python
1, 2, 3
```

但是注意：`map()` 回傳的是一個 map object，不是 list。
如果要看到內容，通常會寫：

```python
list(map(...))
```


````python
data = ['1', '2', '3', '4', '5']

result = map(int, data)
print(result)

# 轉成 list 才看得到內容
print(list(result))
````

## 2. 競賽最常見寫法：讀入一整行數字

假設輸入是一行：

```text
10 20 30 40
```

我們想把它變成整數串列：

```python
[10, 20, 30, 40]
```

常見寫法：

```python
numbers = list(map(int, input().split()))
```

拆開來看：

```python
input()          # 讀入一整行字串，例如 '10 20 30 40'
.split()         # 用空白切開，變成 ['10', '20', '30', '40']
map(int, ...)    # 每個字串都轉成 int
list(...)        # 轉成真正的 list
```


````python
line = '10 20 30 40'

step1 = line.split()
print('split 後：', step1)

step2 = map(int, step1)
print('map 後：', step2)

step3 = list(step2)
print('list 後：', step3)
````

### 實際競賽寫法

如果題目給你一行很多整數，就可以這樣寫：

```python
numbers = list(map(int, input().split()))
```

在 Notebook 裡不方便直接要求學生輸入很多次，所以這裡用字串模擬。


````python
# 模擬 input() 讀到的資料
line = '5 1 2 3 4 5'

numbers = list(map(int, line.split()))

print(numbers)
print('第一個數字 n =', numbers[0])
print('後面的資料 =', numbers[1:])
````

## 3. `a, b = map(int, input().split())`

如果一行剛好有兩個數字，例如：

```text
3 5
```

可以直接拆成兩個變數：

```python
a, b = map(int, input().split())
```

這裡不用 `list()` 也可以，因為 Python 會直接把 map object 裡的值依序拆給 `a` 和 `b`。

但是數量要剛好。

- 左邊 2 個變數，右邊就要有 2 個值
- 左邊 3 個變數，右邊就要有 3 個值


````python
line = '3 5'

a, b = map(int, line.split())

print(a)
print(b)
print(a + b)
````

### 常見錯誤：數量不對

例如左邊只有兩個變數，但右邊有三個數字：

```python
a, b = map(int, '1 2 3'.split())
```

會出錯：

```text
ValueError: too many values to unpack
```


````python
# 這格故意示範錯誤，先不要執行也可以
a, b = map(int, '1 2 3'.split())
````

## 4. `map()` 搭配自己寫的函式

`map()` 不一定只能搭配 `int`。

只要是函式都可以。

例如我們寫一個平方函式：

```python
def square(x):
    return x ** 2
```

然後用：

```python
list(map(square, [1, 2, 3, 4, 5]))
```


````python
def square(x):
    return x ** 2

result = list(map(square, [1, 2, 3, 4, 5]))
print(result)
````

## 5. `map()` 搭配 `lambda`

`lambda` 是一種簡短的匿名函式。

例如：

```python
lambda x: x ** 3
```

意思大概等於：

```python
def f(x):
    return x ** 3
```

所以可以寫：

```python
list(map(lambda x: x ** 3, [1, 2, 3, 4, 5]))
```


````python
result = list(map(lambda x: x ** 3, [1, 2, 3, 4, 5]))
print(result)
````

## 6. `map()` 可以同時處理兩個 list

例如：

```python
a = [1, 2, 3]
b = [4, 5, 6]
```

想要做：

```text
1 + 4 = 5
2 + 5 = 7
3 + 6 = 9
```

可以寫：

```python
summed = list(map(lambda x, y: x + y, a, b))
```


````python
a = [1, 2, 3]
b = [4, 5, 6]

summed = list(map(lambda x, y: x + y, a, b))
print(summed)
````

### 如果兩個 list 長度不同呢？

`map()` 會以比較短的那個為主。

例如：

```python
a = [1, 2, 3, 4]
b = [10, 20]
```

只會算：

```text
1 + 10
2 + 20
```


````python
a = [1, 2, 3, 4]
b = [10, 20]

result = list(map(lambda x, y: x + y, a, b))
print(result)
````

## 7. `map()` 是 lazy evaluation

這點很重要。

`map()` 不會馬上把所有結果算出來。
它會等你真的需要資料時，才一個一個算。

所以：

```python
r = map(int, ['1', '2', '3'])
```

這時候還沒有變成 list。

要用：

```python
list(r)
```

才會真的全部取出。


````python
r = map(int, ['1', '2', '3'])

print('第一次轉 list：', list(r))
print('第二次轉 list：', list(r))
````

上面第二次是空的，因為 `map` 像一條輸送帶。

資料走過一次就沒了。

所以如果資料之後還要重複使用，請先轉成 list：

```python
nums = list(map(int, input().split()))
```


## 8. `map()` vs list comprehension

下面兩種寫法意思很接近：

```python
list(map(int, data))
```

```python
[int(x) for x in data]
```

通常：

- `map(int, data)` 適合單純轉型，例如字串轉整數
- list comprehension 適合邏輯比較複雜的情況

競賽中兩種都常見。


````python
data = ['1', '2', '3', '4', '5']

method1 = list(map(int, data))
method2 = [int(x) for x in data]

print(method1)
print(method2)
````

### 小型效能比較

注意：效能會受到電腦、Python 版本、資料大小影響。

這裡不是要學生死背誰一定比較快，而是理解：

- `map(int, ...)` 對單純轉型通常很常用
- list comprehension 可讀性高、彈性大


````python
from time import perf_counter

N = 300_000
data = [str(i) for i in range(N)]

start = perf_counter()
nums1 = list(map(int, data))
end = perf_counter()
print(f'list(map(int, data))：{end - start:.6f} 秒')

start = perf_counter()
nums2 = [int(x) for x in data]
end = perf_counter()
print(f'[int(x) for x in data]：{end - start:.6f} 秒')

print(nums1[:5], nums2[:5])
````

## 9. 競賽常用模板整理

### 一行多個整數，存成 list

```python
nums = list(map(int, input().split()))
```

### 一行兩個整數，拆成兩個變數

```python
a, b = map(int, input().split())
```

### 一行三個整數

```python
a, b, c = map(int, input().split())
```

### 多行輸入，每行一個整數

```python
n = int(input())
arr = []
for _ in range(n):
    arr.append(int(input()))
```

### 大量輸入時搭配 `sys.stdin.read()`

```python
import sys
nums = list(map(int, sys.stdin.read().split()))
```

## 11. 課堂練習 2

輸入兩行數字，請把對應位置相加。

範例輸入：

```text
1 2 3
4 5 6
```

範例輸出：

```text
5 7 9
```


````python
# 參考答案：Notebook 版用字串模擬兩行 input
line1 = '1 2 3'
line2 = '4 5 6'

a = list(map(int, line1.split()))
b = list(map(int, line2.split()))

ans = list(map(lambda x, y: x + y, a, b))
print(*ans)
````

如果是在比賽平台，寫法會是：

```python
a = list(map(int, input().split()))
b = list(map(int, input().split()))

ans = list(map(lambda x, y: x + y, a, b))
print(*ans)
```


## 12. 重點

`map()` 的核心觀念：

> 把同一個函式，套用到一串資料的每一個元素。

競賽最常用：

```python
nums = list(map(int, input().split()))
```

意思是：

> 讀一行 → 用空白切開 → 每個都轉成整數 → 存成 list。

如果只要兩個數字：

```python
a, b = map(int, input().split())
```

如果資料很多：

```python
import sys
nums = list(map(int, sys.stdin.read().split()))
```
