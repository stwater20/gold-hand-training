## 1. f-string 字串格式化

在 Python 中，`f-string` 可以把變數直接放進字串裡。

格式：

```python
f"文字 {變數} 文字"
```

常用情境：

- 輸出姓名、分數、排名
- 控制小數位數
- 對齊欄位
- 補零輸出

這在比賽題目中很常見，尤其是需要固定格式輸出的題目。

````python
name = "Alice"
score = 95.678

print(f"Name: {name} 好帥")
print(f"Score: {score}")
````

### 1.1 對齊輸出

`{name:>10}` 的意思是：

- `>`：靠右對齊
- `10`：總寬度為 10 個字元

如果字串長度不足 10，左邊會自動補空白。

````python
name = "Alice"

print(f"|{name:>10}|")  # 右對齊，寬度 10
print(f"|{name:<10}|")  # 左對齊，寬度 10
print(f"|{name:^10}|")  # 置中對齊，寬度 10
````

### 1.2 控制小數位數

`{score:.2f}` 的意思是：

- `f`：以小數格式輸出
- `.2`：保留 2 位小數

注意：這會進行四捨五入。

````python
score = 95.678

print(f"{score:.2f}")
print(f"{score:.1f}")
print(f"{score:.0f}")
````

### 1.3 補零輸出

`{score:08.2f}` 的意思是：

- 總寬度為 8
- 保留 2 位小數
- 不足的地方用 `0` 補齊

這種格式常出現在編號、金額、時間輸出。

````python
score = 95.678

print(f"{score:08.2f}")

number = 7
print(f"{number:03d}")  # 007
````

### 小練習

請輸出以下格式：

```text
Student:        Bob
Score:        87.50
ID:             005
```

````python
student = "Bob"
score = 87.5
student_id = 5

print(f"Student: {student:>10}")
print(f"Score:   {score:>10.2f}")
print(f"ID:      {student_id:>1003d}") 
````

## 2. 比較運算

Python 支援一般的比較運算：

| 運算子 | 意義 | 範例 |
|---|---|---|
| `>` | 大於 | `A > 10` |
| `<` | 小於 | `A < 10` |
| `>=` | 大於等於 | `A >= 10` |
| `<=` | 小於等於 | `A <= 10` |
| `==` | 等於 | `A == 10` |
| `!=` | 不等於 | `A != 10` |

比較運算的結果會是布林值：`True` 或 `False`。

````python
A = 50

print(A > 10)
print(A < 10)
print(A == 50)
print(A != 50)
````

### 2.1 傳統寫法 vs Python 寫法

傳統寫法通常會寫成：

```python
A > 1000 and A < 10000
```

Python 可以寫得更像數學式：

```python
1000 < A < 10000
```

這叫做 **連續比較 chained comparison**。

````python
A = 5000

print(A > 1000 and A < 10000)
print(1000 < A < 10000)
````

### 2.2 多個變數的連續比較

例如：

```python
1 < A < 3 < B < 10
```

等價於：

```python
1 < A and A < 3 and 3 < B and B < 10
```

注意：它不是只有檢查 `A` 和 `B`，而是每一段都要成立。

````python
A = 2
B = 5

print(1 < A < 3 < B < 10)

A = 4
B = 5
print(1 < A < 3 < B < 10)
````

有時候寫成傳統寫法反而比較清楚：

```python
1 < A and A < 3 and 3 < B and B < 10
```

## 3. 字串比較

Python 可以直接比較字串大小。

字串比較的規則大致如下：

1. 從第一個字元開始比較。
2. 每個字元根據 Unicode 編碼大小比較。
3. 如果前面都一樣，但是其中一個字串比較短，較短的字串比較小。

這種比較方式很像字典排序。

````python
print("abc" == "abc")
print("abc" > "Abc")
print("abc" < "b")
print("abc" < "abcd")
print("English" < "中文")
````

### 3.1 為什麼 `"abc" > "Abc"` 是 True？

因為 Python 比較的是 Unicode 編碼。

可以用 `ord()` 查看字元的編碼：

````python
print(ord("a"))
print(ord("A"))
print("abc" > "Abc")
````

### 3.2 字串排序範例

如果題目要求姓名排序、代碼排序、字典序比較，就可能會用到字串比較。

````python
words = ["banana", "Apple", "apple", "中文", "English", "abc", "abcd"]

print(sorted(words))
````

在比賽題目中，如果題目說「字典序」或「lexicographical order」，通常就可以直接使用 Python 的字串比較或 `sorted()`。

但要注意大小寫：`A` 和 `a` 的排序結果不一樣。

## 4. Packing 與 Unpacking

Python 很常用到 packing 和 unpacking。

- **Unpacking**：把一組資料拆成多個變數。
- **Packing**：把多個資料收集成一個 list 或 tuple。

這在輸入處理、交換變數、函式參數中非常常見。

### 4.1 Unpacking 基本用法

````python
A, B = 1, 2
print(A)
print(B)
````

### 4.2 交換變數

Python 可以不用第三個暫存變數就交換兩個值。

````python
A = 1
B = 2

A, B = B, A


print(A)
print(B)
````

### 4.3 巢狀 unpacking

如果右邊的資料有巢狀結構，左邊也可以對應拆開。

````python
X, (Y, Z) = 1, (2, 3)

print(Y, Z)
print(X)
print(Y)
print(Z)
````

### 4.4 Packing：使用 `*` 收集剩下的資料

`*z` 會把剩下的資料收集成 list。

````python
x, y, *z = 1, 2, 3, 4, 5

print(x)
print(y)
print(z)
````

````python
x, *y, z = 1, 2, 3, 4, 5

print(x)
print(y)
print(z)
````

### 4.5 在輸入中的常見用法

比賽中常見：

```python
a, b = map(int, input().split())
```

意思是：

1. `input()` 讀一行字串。
2. `.split()` 用空白切開。
3. `map(int, ...)` 把每個字串轉成整數。
4. `a, b = ...` 把兩個整數分別放進 `a` 和 `b`。

````python
line = "10 20"
a, b = map(int, line.split())

print(a)
print(b)
print(a + b)
````

## 5. Shallow Copy

這是 Python 初學者非常容易錯的地方。

先看一個例子：

````python
A = [1, 2, 3, 4]
B = A
print(B)
B[0] = 100

print(A)
print(B)
````

### 5.1 為什麼 A 也被改掉？

因為：

```python
B = A
```

不是複製一份新的 list，而是讓 `B` 指向跟 `A` 同一個 list。

所以改 `B[0]`，其實就是改同一份資料。

### 5.2 一維 list 的複製

如果想複製一份新的 list，可以使用：

```python
B = A.copy()
```

或：

```python
B = A[:]
```

````python
A = [1, 2, 3, 4]
B = A.copy()

B[0] = 100

print(A)
print(B)
````

### 5.3 但是 `.copy()` 只做淺層複製

如果 list 裡面還有 list，也就是二維 list，`.copy()` 只會複製外層，內層 list 仍然是共用的。

````python
A = [1, 2, [3, 4]]
B = A.copy()

B[0] = 100
B[2][0] = 999

print(A)
print(B)
````

觀察結果：

- `B[0] = 100` 不會影響 `A[0]`。
- `B[2][0] = 999` 會影響 `A[2][0]`。

原因是內層的 `[3, 4]` 還是同一份資料。

## 6. Deep Copy

如果想完整複製巢狀資料結構，可以使用 `copy.deepcopy()`。

````python
import copy

A = [1, 2, [3, 4]]
B = copy.deepcopy(A)

B[0] = 100
B[2][0] = 999

print(A)
print(B)
````

### 6.1 Shallow Copy vs Deep Copy 比較

| 方法 | 外層是否複製 | 內層是否複製 | 適合情境 |
|---|---|---|---|
| `B = A` | 否 | 否 | 只是想取另一個名字 |
| `B = A.copy()` | 是 | 否 | 一維 list |
| `B = A[:]` | 是 | 否 | 一維 list |
| `copy.deepcopy(A)` | 是 | 是 | 二維以上、巢狀資料 |

競賽中如果資料是二維陣列、地圖、棋盤、迷宮，要特別小心。

## 7. 常見陷阱：二維串列初始化

這是非常常見的錯誤寫法：

```python
visited = [[False] * C] * R
```

看起來像是建立 R 列 C 欄的二維陣列，但實際上每一列會指向同一個 list。

````python
R = 3
C = 4

visited = [[False] * C] * R
visited[0][0] = True

for row in visited:
    print(row)
````

### 7.1 為什麼三列都被改？

因為：

```python
[[False] * C] * R
```

是把同一列重複 R 次，不是建立 R 個獨立的列。

所以改第一列的第一格，其他列的第一格也會一起變。

### 7.2 正確寫法

應該用 list comprehension：

````python
R = 3
C = 4

visited = [[False] * C for _ in range(R)]
visited[0][0] = True

for row in visited:
    print(row)
````

### 7.3 在 Dijkstra / BFS / DFS 中的影響

如果你在圖論題目中使用：

```python
visited = [[False] * C] * R
```

可能會導致：

- 你只拜訪一個格子，整欄或多列都被標記成拜訪過。
- BFS / DFS 搜尋範圍錯誤。
- Dijkstra 最短路徑結果錯誤。
- 程式不一定會 Runtime Error，但答案會 Wrong Answer。

這種錯誤很難除錯，所以一定要養成正確寫法。

````python
# 錯誤寫法
R = 3
C = 4
visited_wrong = [[False] * C] * R

# 正確寫法
visited_right = [[False] * C for _ in range(R)]

print("錯誤寫法：")
for row in visited_wrong:
    print(id(row), row)

print("正確寫法：")
for row in visited_right:
    print(id(row), row)
````

可以觀察 `id(row)`：

- 錯誤寫法中，每一列的 id 相同，代表其實是同一個 list。
- 正確寫法中，每一列的 id 不同，代表是不同的 list。
