# Python `set`

1. `set` 的特性：不重複、無序、查找 O(1)
2. 集合運算：交集、聯集、差集
3. 競賽常見用法：去重複、快速判斷「有沒有出現過」


````python
# 建立 set：重複的元素自動消失
s = set([1, 2, 2, 3, 3, 3])
print(s)

words = "apple banana apple cherry banana".split()
print(set(words))
print(len(set(words)))     # 不同單字的數量
````

## 1. `in` 查找：`set` 是 O(1)，`list` 是 O(n)

（詳細測速見 `6_python_performance_tips_demo.ipynb`）


````python
from time import perf_counter

large_value = 1000000
search_value = 999999

data_list = list(range(large_value))
data_set = set(data_list)

start = perf_counter()
print(search_value in data_list)
print(f"list 查找: {perf_counter() - start:.6f} 秒")

start = perf_counter()
print(search_value in data_set)
print(f"set  查找: {perf_counter() - start:.6f} 秒")
````

## 2. 集合運算


````python
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(a | b)   # 聯集
print(a & b)   # 交集
print(a - b)   # 差集：在 a 不在 b
print(a ^ b)   # 對稱差：只出現在其中一邊

s = set()
s.add(10)
s.add(10)      # 加第二次沒有效果
s.discard(99)  # 移除不存在的元素也不會錯（remove 會 KeyError）
print(s)
````

## 3. 模擬試題：不同單字數

【題目】第一列是整數 $n$（此處僅為干擾），第二列有若干個以空白隔開的單字，
請輸出「不同單字」的數量。

【範例輸入】

```text
5
dog cat dog bird cat
```

【範例輸出】

```text
3
```


````python
sample = "5\ndog cat dog bird cat"
inf = sample.split("\n")

n_ = int(inf.pop(0))
ch = set(inf.pop(0).strip().split())
ans = len(ch)
print(ans)
````

## 4. 練習題

輸入兩列名單，輸出「兩列都出現」的名字（依字母排序）。

輸入：

```text
Amy Bob Cindy
Bob David Amy
```

期望輸出：

```text
Amy Bob
```


````python
# 練習作答區
line1 = "Amy Bob Cindy".split()
line2 = "Bob David Amy".split()
````

### 參考答案

````python
line1 = "Amy Bob Cindy".split()
line2 = "Bob David Amy".split()
print(*sorted(set(line1) & set(line2)))
````
