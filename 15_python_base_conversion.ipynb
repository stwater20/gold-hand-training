# 進位轉換

1. 其他進位轉換到 10 進位：`int(字串, 進位)`
2. 10 進位轉換到 2、8 進位：`bin()`、`oct()`、`format()`、f-string
3. 10 進位轉換到 16 進位：`hex()`、`format()`、f-string


## 1. 其他進位轉換到 10 進位

`int(字串, base)` 可以把任何 2～36 進位的字串轉成 10 進位整數。


````python
print(int("1110", 2))    # 2進位 → 14
print(int("0b1110", 2))  # 有前綴也可以
print(int("70", 8))      # 8進位 → 56
print(int("ff", 16))     # 16進位 → 255
print(int("0xff", 16))
print(int("z", 36))      # 36進位 → 35
````

## 2. 10 進位轉換到 2、8 進位

三種寫法：`bin()` / `oct()`、`format()`、f-string。


````python
# 10進位轉2進位
print(bin(3))              # '0b11'
print(bin(-10))            # '-0b1010'
print(format(14, '#b'))    # '0b1110'
print(format(14, 'b'))     # '1110'（沒有前綴）
print(f"{14:#b}")          # '0b1110'
print(f"{14:b}")           # '1110'
````

````python
# 10進位轉8進位
print(oct(8))              # '0o10'
print(oct(-56))            # '-0o70'
print(format(8, '#o'))     # '0o10'
print(format(8, 'o'))      # '10'
print(f"{8:#o}")
print(f"{8:o}")
````

## 3. 10 進位轉換到 16 進位


````python
print(hex(255))            # '0xff'
print(hex(-42))            # '-0x2a'
print(format(255, '#x'))   # '0xff'
print(format(255, 'X'))    # 'FF'（大寫）
print(format(255, 'x'))    # 'ff'
print(f"{255:#x}")
print(f"{255:x}")
print(f"{255:X}")
````

## 4. 補零對齊：競賽輸出常用


````python
# 轉 2 進位並補零到 8 位
print(f"{14:08b}")     # 00001110
# 轉 16 進位並補零到 4 位
print(f"{255:04x}")    # 00ff
````

## 5. 任意進位互轉

Python 沒有內建「10 進位轉任意進位」的函式（只有 2、8、16），
其他進位要自己寫短除法。


````python
def to_base(n, base):
    if n == 0:
        return "0"
    digits = "0123456789abcdefghijklmnopqrstuvwxyz"
    result = []
    while n > 0:
        result.append(digits[n % base])
        n //= base
    return "".join(reversed(result))

print(to_base(100, 3))                 # 10201
print(int(to_base(100, 3), 3))         # 驗證：轉回來是 100
print(to_base(255, 5))
````
