# Python `zip()` 



1. `zip(a, b)` 把兩個序列「一個配一個」配對
2. `zip()` 遇到長度不同的序列會停在最短的那個
3. `zip(*資料)` 可以做「反向拆開」或「轉置」
4. 用 `zip()` 寫矩陣相乘

> `zip()` 本身不會直接產生 list，它會產生一個 iterator，所以通常會用 `list(zip(...))` 看結果。

## 1. `zip(a, b)`：把兩個 list 配對

想像有兩排資料：

```text
a = [1, 2, 3]
b = [4, 5, 6]
```

`zip(a, b)` 會變成：

```text
(1, 4), (2, 5), (3, 6)
```

也就是第 0 個配第 0 個，第 1 個配第 1 個，第 2 個配第 2 個。

````python
a = [1, 2, 3]
b = [4, 5, 6]

zip1 = zip(a, b)
print(list(zip1))
````

## 2. 長度不同時：只會配到最短的那個

如果：

```python
a = [1, 2, 3]
c = [4, 5, 6, 7, 8]
```

`zip(a, c)` 只會產生三組，因為 `a` 只有三個元素。

````python
a = [1, 2, 3]
c = [4, 5, 6, 7, 8]

zip2 = zip(a, c)
print(list(zip2))
````

## 3. 注意：`zip()` 是 iterator，用完就沒了


下面同一個 `z` 印兩次，第二次會變成空的。

````python
a = [1, 2, 3]
b = [4, 5, 6]

z = zip(a, b)
print(list(z))
print(list(z))  # 第二次已經沒有資料了
````

所以如果資料後面還要重複使用，可以先轉成 list：

````python
a = [1, 2, 3]
b = [4, 5, 6]

pairs = list(zip(a, b))
print(pairs)
print(pairs)
````

## 4. `x, y = zip(*pairs)`：把配好的資料拆回來

假設我們有：

```python
pairs = [(1, 4), (2, 5), (3, 6)]
```

使用：

```python
x, y = zip(*pairs)
```

可以拆回：

```python
x = (1, 2, 3)
y = (4, 5, 6)
```

這裡的 `*pairs` 是 unpack，意思是把 list 裡面的元素展開。

````python
pairs = [(1, 4), (2, 5), (3, 6)]

x, y = zip(*pairs)
print(x)
print(y)
````

## 5. `zip(*matrix)`：矩陣轉置

矩陣：

```text
A = [
    [1, 2, 3],
    [4, 5, 6]
]
```

原本是 2 列 3 欄。

`zip(*A)` 會把「欄」拿出來，變成：

```text
(1, 4), (2, 5), (3, 6)
```

這也叫做矩陣轉置 transpose。

````python
A = [
    [1, 2, 3],
    [4, 5, 6]
]

transposed = list(zip(*A))
print(transposed)
````

如果想要轉成 list of list，可以這樣：

````python
A = [
    [1, 2, 3],
    [4, 5, 6]
]

transposed = [list(col) for col in zip(*A)]
print(transposed)
````

## 6. `zip()` 常見用途：兩個 list 對應運算

例如兩個 list 對應相加：

```text
a = [1, 2, 3]
b = [4, 5, 6]
結果 = [5, 7, 9]
```

````python
a = [1, 2, 3]
b = [4, 5, 6]

result = []
for x, y in zip(a, b):
    result.append(x + y)

print(result)
````

也可以寫成 list comprehension：

````python
a = [1, 2, 3]
b = [4, 5, 6]

result = [x + y for x, y in zip(a, b)]
print(result)
````

## 7. 內積 Dot Product

矩陣相乘前，先理解「內積」。

例如：

```text
row = [1, 2]
col = [7, 9]
```

內積就是：

```text
1 * 7 + 2 * 9 = 25
```

用 `zip(row, col)` 可以把兩個向量一一配對：

```text
(1, 7), (2, 9)
```

````python
row = [1, 2]
col = [7, 9]

dot_product = sum(a * b for a, b in zip(row, col))
print(dot_product)
````

## 8. 矩陣相乘：A × B

投影片範例：

```text
A = [[1, 2],
     [3, 4],
     [5, 6]]

B = [[7, 8],
     [9, 10]]
```

A 是 3×2，B 是 2×2，所以結果會是 3×2。

計算規則：

> A 的每一列 row，分別去乘 B 的每一欄 col。

第一格：

```text
[1, 2] · [7, 9] = 1*7 + 2*9 = 25
```

第二格：

```text
[1, 2] · [8, 10] = 1*8 + 2*10 = 28
```

````python
A = [
    [1, 2],
    [3, 4],
    [5, 6]
]

B = [
    [7, 8],
    [9, 10]
]

# 先把 B 的欄取出來
B_cols = list(zip(*B))
print(B_cols)
````

`B_cols` 變成：

```python
[(7, 9), (8, 10)]
```

也就是 B 的第一欄和第二欄。

````python
A = [
    [1, 2],
    [3, 4],
    [5, 6]
]

B = [
    [7, 8],
    [9, 10]
]

B_cols = list(zip(*B))

result = []
for row_A in A:
    row_result = []
    for col_B in B_cols:
        dot_product = sum(a * b for a, b in zip(row_A, col_B))
        row_result.append(dot_product)
    result.append(row_result)

for row in result:
    print(row)
````

## 9. 矩陣相乘：一行版本

上面那段可以濃縮成 list comprehension：

````python
A = [
    [1, 2],
    [3, 4],
    [5, 6]
]

B = [
    [7, 8],
    [9, 10]
]

result = [
    [sum(a * b for a, b in zip(row_A, col_B)) for col_B in zip(*B)]
    for row_A in A
]

for row in result:
    print(row)
````

## 10. 矩陣相乘函式版

這是比較適合放進正式程式的版本。

````python
def matrix_multiply(A, B):
    # A: m x n
    # B: n x p
    # result: m x p
    B_cols = list(zip(*B))

    result = []
    for row_A in A:
        row_result = []
        for col_B in B_cols:
            value = sum(a * b for a, b in zip(row_A, col_B))
            row_result.append(value)
        result.append(row_result)
    return result

A = [
    [1, 2],
    [3, 4],
    [5, 6]
]

B = [
    [7, 8],
    [9, 10]
]

result = matrix_multiply(A, B)
for row in result:
    print(row)
````

## 11. 加上維度檢查

矩陣相乘需要：

```text
A 的欄數 == B 的列數
```

例如 A 是 3×2，B 是 2×2，可以乘。

但如果 A 是 3×2，B 是 3×2，就不能直接做 A × B。

````python
def matrix_multiply_checked(A, B):
    if len(A) == 0 or len(B) == 0:
        raise ValueError("A 和 B 都不能是空矩陣")

    a_cols = len(A[0])
    b_rows = len(B)

    if a_cols != b_rows:
        raise ValueError(f"維度不符合：A 的欄數是 {a_cols}，但 B 的列數是 {b_rows}")

    B_cols = list(zip(*B))
    return [
        [sum(a * b for a, b in zip(row_A, col_B)) for col_B in B_cols]
        for row_A in A
    ]

A = [[1, 2], [3, 4], [5, 6]]
B = [[7, 8], [9, 10]]

print(matrix_multiply_checked(A, B))
````
