# 佇列 Queue、堆疊 Stack 與 `deque`

1. Queue（FIFO）的三種實作：`queue.Queue`、`list`、`collections.deque`
2. Stack（LIFO）的實作：`list`、`LifoQueue`、`deque`
3. Priority Queue 優先佇列
4. 各種實作的速度比較：**比賽請用 `deque` 和 `list`**


## 1. 佇列 Queue

具有 FIFO（First In, First Out）性質的有序串列：
插入發生在 Rear（尾）端，刪除發生在 Front（前）端。


### 實作一：`queue.Queue`

常用方法：`put()`、`get()`、`qsize()`、`empty()`

> `queue.Queue` 是為「多執行緒」設計的，有鎖的成本，**比賽不建議使用**。


````python
from queue import Queue

n_ = 10
q = Queue()
for i in range(n_):
    q.put(i)
print(q.qsize())
print(q.empty())
while not q.empty():
    print(q.get(), end=" ")
print()
````

### 實作二：`list`

常用方法：`append()`、`del q[0]`、`len(q)`

> 注意：`del q[0]` 是 O(n)，資料量大時很慢。


````python
n_ = 10
q = []
for i in range(n_):
    q.append(i)
print(q)
print(len(q))
for i in range(n_):
    print(q[0], end=" ")
    del q[0]
print()
print(q)
````

### 實作三：`collections.deque`（推薦）

常用方法：`append()`、`appendleft()`、`pop()`、`popleft()`、`len()`

兩端插入與刪除都是 O(1)。


````python
from collections import deque

n_ = 10
q = deque()
for i in range(n_):
    q.append(i)
print(len(q))
while q:
    print(q.popleft(), end=" ")
print()
````

## 2. `deque` 常用操作整理


````python
from collections import deque

q = deque()
q.append(1)        # 右端加入
q.append(2)
q.appendleft(0)    # 左端加入
print(q)
print(q[0])        # 查看左端
print(q[-1])       # 查看右端
print(q.pop())     # 右端移除
print(q.popleft()) # 左端移除
print(q)
````

````python
# 錯誤作法：list 在開頭插入是 O(n)
# mylist.insert(0, x)

# 正確做法：deque 左端加入是 O(1)
from collections import deque
from time import perf_counter

n_ = 100000

start = perf_counter()
mylist = []
for x in range(n_):
    mylist.insert(0, x)
print(f"list.insert(0, x):    {perf_counter() - start:.4f} 秒")

start = perf_counter()
dq = deque()
for x in range(n_):
    dq.appendleft(x)
print(f"deque.appendleft(x):  {perf_counter() - start:.4f} 秒")
````

## 3. Queue、list、deque 的速度比較


````python
from queue import Queue
from collections import deque
from time import perf_counter

n_ = 200000

start = perf_counter()
q = Queue()
for i in range(n_):
    q.put(i)
for i in range(n_):
    q.get()
print(f"queue.Queue : {perf_counter() - start:.4f} 秒")

start = perf_counter()
l = []
for i in range(n_):
    l.append(i)
for i in range(n_):
    del l[0]
print(f"list        : {perf_counter() - start:.4f} 秒")

start = perf_counter()
dq = deque()
for i in range(n_):
    dq.append(i)
for i in range(n_):
    dq.popleft()
print(f"deque       : {perf_counter() - start:.4f} 秒")
````

## 4. Priority Queue 優先佇列

特點：優先權最高的資料（最小值）始終位於佇列頂端。

基本操作：

```python
pq = queue.PriorityQueue()
pq.put([priority, value])
pq.get()
pq.empty()
pq.qsize()
```

> 同樣因為鎖的成本，比賽建議直接用 `heapq`（見 `27_python_heapq_demo.ipynb`）。


````python
from queue import PriorityQueue

pq = PriorityQueue()
pq.put(7)
pq.put(1)
pq.put(5)
print(pq.qsize())
while not pq.empty():
    print(pq.get(), end="  ")   # 由小到大出來
print()

# priority + value 的形式
pq = PriorityQueue()
pq.put([2, "寫作業"])
pq.put([1, "吃飯"])
pq.put([3, "睡覺"])
while not pq.empty():
    print(pq.get())
````

## 5. 堆疊 Stack

將資料依序往上疊（Push），取出時從最上面拿（Pop）：LIFO（Last In, First Out）。


### 實作一：`LifoQueue`

常用方法：`put()`、`get()`、`empty()`、`qsize()`


````python
from queue import LifoQueue

st = LifoQueue(maxsize=100)
st.put("Yiming")
st.put(98)
print(st.qsize())
print(st.get())    # 後進先出 → 98
print(st.get())
````

### 實作二：`list` 或 `deque`（推薦）


````python
# list 當 stack
stack = []
stack.append("Yiming")
stack.append(98)
print(len(stack))
print(stack.pop())
print(stack.pop())

# deque 當 stack
from collections import deque
stack = deque()
stack.append("Yiming")
stack.append(98)
print(stack.pop())
print(stack.pop())
````

## 6. Stack 三種實作的速度比較


````python
from queue import LifoQueue
from collections import deque
from time import perf_counter

n_ = 500000

start = perf_counter()
st = LifoQueue(maxsize=n_)
for i in range(n_):
    st.put(i)
for i in range(n_):
    st.get()
print(f"LifoQueue : {perf_counter() - start:.4f} 秒")

start = perf_counter()
st = []
for i in range(n_):
    st.append(i)
for i in range(n_):
    st.pop()
print(f"list      : {perf_counter() - start:.4f} 秒")

start = perf_counter()
st = deque()
for i in range(n_):
    st.append(i)
for i in range(n_):
    st.pop()
print(f"deque     : {perf_counter() - start:.4f} 秒")
````

**結論**：

| 用途 | 比賽推薦 | 避免 |
|---|---|---|
| Queue | `deque`（`popleft`） | `queue.Queue`、`list` 的 `del q[0]` |
| Stack | `list` 或 `deque`（`pop`） | `LifoQueue` |
| Priority Queue | `heapq` | `PriorityQueue` |


## 7. 練習題：括號配對

輸入一個只含 `()[]{}` 的字串，判斷括號是否合法配對。

輸入：`([]{()})` → 輸出 `True`；輸入 `([)]` → 輸出 `False`


````python
# 練習作答區
s1 = "([]{()})"
s2 = "([)]"

# 提示：用 stack，遇到左括號 push，遇到右括號檢查 top 是否配對
````

### 參考答案

````python
def valid(s):
    pair = {')': '(', ']': '[', '}': '{'}
    stack = []
    for ch in s:
        if ch in "([{":
            stack.append(ch)
        else:
            if not stack or stack.pop() != pair[ch]:
                return False
    return not stack

print(valid("([]{()})"))
print(valid("([)]"))
````
