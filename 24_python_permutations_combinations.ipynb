## 1. 排列 `permutations`


````python
from itertools import permutations

letters = "SCVS"
a = permutations(letters, 3)     # 取 3 個的排列
y = [''.join(i) for i in a]
print(y)
print(len(y))
````

## 2. 注意事項一：依「位置順序」排列，不是字典序

`permutations` 按輸入元素的位置順序進行排列，並不是按字元的字典序輸出。


````python
from itertools import permutations

s = ['b', 'a', 'c']
t = ["".join(e) for e in permutations(s)]
print(t)    # 從 'bac' 開始，不是 'abc'

# 想要字典序：先排序再 permutations
t2 = ["".join(e) for e in permutations(sorted(s))]
print(t2)
````

## 3. 注意事項二：相同元素被視為「不同」

如果有相同的元素，不同位置的元素被認為不同 → 會產生重複的排列。


````python
from itertools import permutations

s = ['A', 'A', 'B']
t = ["".join(e) for e in permutations(s)]
print(t)              # 'AAB' 出現兩次

# 去除重複：套 set
print(sorted(set(t)))
````

## 4. 組合 `combinations`


````python
from itertools import combinations

letters = "SCVS"
a = combinations(letters, 3)
y = [''.join(i) for i in a]
print(y)
````

## 5. 注意事項：組合一樣把相同元素視為不同


````python
from itertools import combinations

s = ['1', '1', '3', '2']
t = combinations(s, 2)
r = ["".join(e) for e in t]
print(" ".join(r))     # '11' '13' '12' ... 有重複意義的組合

# 先去重、排序，再取組合
t2 = sorted(set(s))
u = combinations(t2, 2)
v = ["".join(e) for e in u]
print(" ".join(v))
````

## 6. 用二進位表示子集合

**已知**：一個子集合可以對應到一個二進位數。

假設有 3 個元素 $\{A, B, C\}$，用 3 位元二進位數表示選擇的組合：

- `000`：沒有選擇任何元素
- `001`：選擇了 C
- `010`：選擇了 B
- `101`：選擇了 A 和 C
- `111`：全選

所以「列舉所有子集合」= 把 $0$ 到 $2^n - 1$ 全部走一遍。


````python
a = [1, 2, 3, 4, 5, 6]
n_ = 4          # 只看前 4 個元素的所有子集合

for i in range(1 << n_):        # 1 << n 就是 2**n
    print('{', end='')
    for j in range(n_):
        if i & (1 << j):        # 第 j 個位元是 1 → 選了 a[j]
            print(a[j], end=',')
    print('}', end=' ')
print()
````

## 7. 如何判斷二進位數字中 1 的個數？

技巧：`k & (k - 1)` 會**消除最右邊的 1**。

假設 $k = 1101_2 = 13$：

1. `1101 & 1100` → `1100`（消除最右邊的 1）
2. `1100 & 1011` → `1000`
3. `1000 & 0111` → `0000`

共進行 3 次 → 有 3 個 1。


````python
def count_ones(k):
    cnt = 0
    while k > 0:
        k = k & (k - 1)
        cnt += 1
    return cnt

print(count_ones(0b1101))     # 3
print(count_ones(255))        # 8
print(bin(100), count_ones(100))

# Python 3.8+ 也可以用內建：
print((13).bit_count() if hasattr(int, "bit_count") else bin(13).count("1"))
````

## 8. $n$ 個數中任意 $m$ 個數的組合（位元版）

走遍 $0 \sim 2^n - 1$，只輸出「1 的個數恰好是 $m$」的子集合。


````python
a = [1, 2, 3, 4, 5, 6, 7]

def print_set(n_, m):
    for i in range(2 ** n_):
        num, k = 0, i
        while k > 0:               # 數 i 裡有幾個 1
            k = k & (k - 1)
            num += 1
        if num == m:               # 恰好選了 m 個
            subset = [a[j] for j in range(n_) if i & (1 << j)]
            print(subset)

print_set(4, 3)
````

## 9. 練習題

有 5 個數 `[3, 7, 1, 8, 4]`，請用「位元列舉」找出：
總和恰好等於 12 的子集合有幾個？

期望輸出：

```text
3
```

（`3+1+8`、`7+1+4`、`8+4`）


````python
# 練習作答區
a = [3, 7, 1, 8, 4]
````

### 參考答案

````python
a = [3, 7, 1, 8, 4]
n_ = len(a)
cnt = 0
for i in range(1 << n_):
    total = sum(a[j] for j in range(n_) if i & (1 << j))
    if total == 12:
        cnt += 1
print(cnt)
````
