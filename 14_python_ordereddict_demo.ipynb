## 1. `OrderedDict` 是什麼？

`OrderedDict` 會記住 key「加入的順序」，走訪時依照加入順序輸出。

> 注意：Python 3.7 之後，一般的 `dict` 也保證保留插入順序。
> 但 `OrderedDict` 還有額外功能（例如 `move_to_end()`、比較時順序也要相同），
> 在競賽中看到舊程式碼用 `OrderedDict` 也要看得懂。


````python
from collections import OrderedDict

ordered_dict = OrderedDict()
ordered_dict['apple'] = 3
ordered_dict['banana'] = 2
ordered_dict['orange'] = 5

for key, value in ordered_dict.items():
    print(key, value)
````

## 2. `OrderedDict` 與 `dict` 的比較


````python
from collections import OrderedDict

# 順序不同的兩個 dict：內容相同就相等
d1 = {'a': 1, 'b': 2}
d2 = {'b': 2, 'a': 1}
print(d1 == d2)          # True

# OrderedDict：順序不同就不相等
od1 = OrderedDict([('a', 1), ('b', 2)])
od2 = OrderedDict([('b', 2), ('a', 1)])
print(od1 == od2)        # False

# move_to_end：把 key 移到最後（LRU Cache 的核心）
od1.move_to_end('a')
print(list(od1.items()))
````

## 3. HackerRank - Word Order

題目連結：<https://www.hackerrank.com/challenges/word-order/problem>

【題目】

給定一堆單字，可能包含重複的單字。輸出「不同單字的數量」，
接著對每個單字輸出其出現次數，輸出順序應該與單字在輸入中**首次出現的順序**相同。

【輸入格式】

第一列包含一個整數 $N$，表示單字的數量。接下來的 $N$ 列，每列一個單字。

【範例輸入】

```text
4
bcdef
abcdefg
bcde
bcdef
```

【範例輸出】

```text
3
2 1 1
```


````python
# 用字串模擬輸入
sample = "4\nbcdef\nabcdefg\nbcde\nbcdef"
data = sample.split("\n")
print(data)
````

````python
from collections import OrderedDict

def word_order(data):
    n = int(data[0])
    word_counts = OrderedDict()
    for i in range(1, n + 1):
        word = data[i].strip()
        word_counts[word] = word_counts.get(word, 0) + 1
    print(len(word_counts))
    print(*word_counts.values())

word_order(data)   # 應該輸出 3 和 2 1 1
````

## 4. 練習題

輸入若干列「學生姓名」，代表到校簽到的順序（同一人只算第一次）。
請依「首次簽到順序」輸出名單。

輸入：`Amy Bob Cindy Bob Amy David`

期望輸出：

```text
Amy Bob Cindy David
```


````python
# 練習作答區
names = "Amy Bob Cindy Bob Amy David".split()

# 提示：OrderedDict 或 dict.fromkeys()
````

### 參考答案

````python
names = "Amy Bob Cindy Bob Amy David".split()

# 解法一：OrderedDict
from collections import OrderedDict
print(*OrderedDict.fromkeys(names))

# 解法二：Python 3.7+ 的 dict 也保留順序
print(*dict.fromkeys(names))
````
