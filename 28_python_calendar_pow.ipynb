# `calendar.isleap` 與三參數 `pow()`

1. `calendar.isleap()`：閏年判斷
2. `pow(x, y, z)`：快速模冪（110 年度試題 K 類型）


## 1. 閏年判斷

規則：西元年能被 4 整除且不能被 100 整除，或能被 400 整除。

自己寫容易漏條件，Python 內建 `calendar.isleap()` 一行搞定。


````python
import calendar

for y in [1900, 2000, 2024, 2025, 2026]:
    print(y, calendar.isleap(y))
````

## 2. 實戰：閏年題

【題目】每列一個西元年份，若為閏年輸出 `a leap year`，否則輸出 `not a leap year`。

【範例輸入】

```text
2000
1900
2024
```


````python
import calendar

sample = "2000\n1900\n2024"
data = sample.split("\n")

def process(data):  # 主要的解題邏輯放在這裡。
    for line in data:
        if calendar.isleap(int(line)):
            print("a leap year")
        else:
            print("not a leap year")

process(data)
````

## 3. `pow(x, y, z)`：快速模冪

計算 $x^y \bmod z$。**千萬不要**寫 `x ** y % z`：
$y$ 很大時 `x ** y` 會先算出天文數字，再取餘數，超慢又吃記憶體。

`pow(x, y, z)` 用「快速冪」演算法，O($\log y$) 就算完。


````python
print(pow(2, 10, 1000))       # 1024 mod 1000 = 24
print(pow(3, 200, 10**9 + 7)) # 大指數也瞬間算完
````

````python
from time import perf_counter

x, y, z = 7, 10**6, 10**9 + 7

start = perf_counter()
r1 = x ** y % z
print(f"x ** y % z   : {r1}, {perf_counter() - start:.4f} 秒")

start = perf_counter()
r2 = pow(x, y, z)
print(f"pow(x, y, z) : {r2}, {perf_counter() - start:.6f} 秒")
````

## 4. 110 年度試題 K 類型

【題目】每 4 列為一組測試資料，前 3 列分別是 $x$、$y$、$z$（第 4 列為空白列），
輸出 $x^y \bmod z$。

【範例輸入】

```text
2
10
1000

3
5
7
```


````python
sample = "2\n10\n1000\n\n3\n5\n7"
data = sample.split("\n")

def process(data):
    l = len(data)
    for i in range(0, l, 4):
        x, y, z = map(int, data[i:i + 3])
        print(pow(x, y, z))

process(data)
````

## 5. 補充：`pow` 的其他用法

`pow(x, -1, z)` 可以算「模反元素」（$x^{-1} \bmod z$，Python 3.8+），
在需要「除法取餘數」的競賽題很好用。


````python
# 3 * 5 = 15 ≡ 1 (mod 7)，所以 3 的模反元素是 5
inv = pow(3, -1, 7)
print(inv)
print(3 * inv % 7)    # 驗證 = 1
````
