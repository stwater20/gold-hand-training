## 1. `Counter` 基本用法

`Counter` 是專門做計數的字典：key 是元素、value 是出現次數。


````python
from collections import Counter

# 直接數字串裡的字元
print(Counter("hello world"))

# 數 list 裡的元素
print(Counter([1, 2, 2, 3, 3, 3]))

# most_common：出現次數排行榜
c = Counter("abracadabra")
print(c.most_common(3))
````

````python
from collections import Counter

# update()：累加計數
animal = Counter()
animal.update("dog cat dog".split())
animal.update("cat dog bird".split())
print(animal)

# 找出現最多次的 key
ans = max(animal, key=animal.get)
print(ans)
````

## 2. 111 年度試題 E：數字出現次數

【題目】第一列是整數 $n$，第二列有 $n$ 個整數。
請依數字由小到大，輸出「數字 出現次數」。

【範例輸入】

```text
5
3 1 2 3 1
```

【範例輸出】

```text
1 2
2 1
3 2
```


````python
from collections import Counter

sample = "5\n3 1 2 3 1"
data = sample.split("\n")

def process(data):  # 主要的解題邏輯放在這裡。
    n_ = int(data[0])
    ans = sorted(Counter(map(int, data[1].split())).items())
    for p in ans:
        print(f"{p[0]} {p[1]}")

process(data)
````

## 3. 111 年度試題 I：字元頻率排序

【題目】給一個字串，統計每個字元出現次數，
依「次數由小到大」輸出，次數相同時字元大的在前。

【範例輸入】

```text
banana
```

【範例輸出】

```text
b 1
n 2
a 3
```


````python
from collections import Counter

data = ["banana"]

def process(data):  # 主要的解題邏輯放在這裡。
    c = Counter(data[0]).items()
    l = sorted([[d[1], -ord(d[0])] for d in c])
    for t in l:
        print(f"{chr(-t[1])} {t[0]}")

process(data)
````

## 4. AtCoder ABC137 C - Green Bin

題目連結：<https://atcoder.jp/contests/abc137/tasks/abc137_c>

【題目說明】把字串 $a$ 的字母重新排列得到的字串稱為 $a$ 的變位字（anagram）。
例如 `greenbin` 是 `beginner` 的變位字。
給定 $N$ 個字串，求有多少對 $(i, j)$（$i < j$）互為變位字。

【範例輸入】

```text
5
abaaaaaaaa
oneplustwo
aaaaaaaaba
twoplusone
aaaabaaaaa
```

【範例輸出】

```text
4
```


````python
sample = "5\nabaaaaaaaa\noneplustwo\naaaaaaaaba\ntwoplusone\naaaabaaaaa"
data = sample.split("\n")
````

## 5. 解題想法

互為變位字 ⟺ 排序後的字串相同。

1. 把每個字串排序當作「標準型」
2. 用 `Counter` 數每種標準型出現幾次
3. 出現 $c$ 次的標準型可以組成 $\binom{c}{2} = \dfrac{c(c-1)}{2}$ 對


````python
from collections import Counter

def count_anagram_pairs(strings):
    sorted_strings = [''.join(sorted(s)) for s in strings]
    count = Counter(sorted_strings)
    total_pairs = 0
    for c in count.values():
        total_pairs += c * (c - 1) // 2
    return total_pairs

N = int(data[0])
strings = [data[i + 1].strip() for i in range(N)]
print(count_anagram_pairs(strings))    # 應輸出 4
````

輸入一段英文句子，輸出「出現次數最多的單字」與其次數
（次數相同時輸出字典序較小者）。

輸入：`the quick the lazy dog the lazy`

期望輸出：

```text
the 3
```


````python
# 練習作答區
sentence = "the quick the lazy dog the lazy"
````

### 參考答案

````python
from collections import Counter

sentence = "the quick the lazy dog the lazy"
c = Counter(sentence.split())
word = min(c, key=lambda w: (-c[w], w))
print(word, c[word])
````
