# `itertools.product`：笛卡兒積

## 1. `product` 基本用法

把多個序列的「所有搭配」全部列出來（笛卡兒積）。


````python
from itertools import product

list1 = [1, 2]
list2 = ['x', 'y']
string1 = 'AB'
result = product(list1, list2, string1)
print(list(result))
````

````python
from itertools import product

# repeat：同一序列重複搭配
print(list(product([0, 1], repeat=3)))   # 3 位元的所有 01 組合
````

## 2. 用 `product` 取代多層巢狀迴圈

三層迴圈：

```python
for a in A:
    for b in B:
        for c in C:
            ...
```

可以寫成一層：

```python
for a, b, c in product(A, B, C):
    ...
```


````python
from itertools import product

A = [1, 2]
B = [10, 20]
C = [100, 200]

for a, b, c in product(A, B, C):
    print(a + b + c, end=" ")
print()
````

## 3. 實戰：每個數列各選一個數

【題目】有 $K$ 個數列與一個整數 $M$。從每個數列中各選一個數，
令 $S$ =（所選數字的總和）$\bmod\ M$，求 $S$ 的最大值。

【輸入格式】

第一列包含兩個整數 $K$ 和 $M$。接下來的 $K$ 列，
每列第一個數為數列長度 $N_i$，接著是 $N_i$ 個正整數。

【範例輸入】

```text
3 100
2 5 4
3 7 8 9
5 5 7 8 9 10
```

【範例輸出】

```text
24
```

（$5 + 9 + 10 = 24$，$24 \bmod 100 = 24$）


````python
from itertools import product

sample = "3 100\n2 5 4\n3 7 8 9\n5 5 7 8 9 10"
data = sample.split("\n")

def solve(data):
    K, M = map(int, data[0].split())
    seqs = []
    for i in range(1, K + 1):
        nums = list(map(int, data[i].split()))
        seqs.append(nums[1:])          # 第一個數是長度，跳過

    best = 0
    for picks in product(*seqs):       # 窮舉所有選法
        best = max(best, sum(picks) % M)
    return best

print(solve(data))
````

## 4. 為什麼敢用窮舉？

約束條件 $1 \le K \le 7$、$1 \le N_i \le 7$，
最多 $7^7 = 823543$ 種選法，Python 也能在時限內跑完。

**競賽心法**：看到「數量很小」的約束（≤ 7、≤ 10、≤ 20），
通常就是在暗示你可以窮舉（`product`、位元列舉或 DFS）。


````python
# 感受一下最壞情況的窮舉量
from itertools import product
from time import perf_counter

seqs = [list(range(1, 8))] * 7        # 7 個數列，每個 7 個元素
start = perf_counter()
best = max(sum(p) % 1000 for p in product(*seqs))
print(f"7^7 = {7**7} 種選法, 最佳值 {best}, 耗時 {perf_counter() - start:.4f} 秒")
````
