# 排序：`sorted()`、`sort()` 與自定義比較函數

1. `sorted()` 與 `list.sort()` 的差別
2. `key` 參數：`len`、`lambda`
3. 雙重排序：`(-x[1], x[0])`
4. `functools.cmp_to_key`：自定義比較函數


## 1. `sorted()`：回傳新 list，原本的不變


````python
L = [3, 1, 4, 2]
print("Sorted list:")
s1 = sorted(L)
print(s1)
print("Reverse sorted list:")
s2 = sorted(L, reverse=True)
print(s2)
print("Original L list:")
print(L)               # 原本的 L 沒變
````

## 2. `key` 參數：指定排序依據


````python
L = ['aaaa', 'bbb', 'cc', 'd']
print("Sorted list:")
print(sorted(L))              # 依字典序
print("Sorted list with key:")
print(sorted(L, key=len))     # 依長度
print("Original L:")
print(L)
````

## 3. `list.sort()`：就地排序，回傳 `None`

**常見錯誤**：`s = L.sort()` 之後 `s` 是 `None`！


````python
L = [1, 3, 4, 2]
s = L.sort(reverse=True)
print("Return Result：")
print(s)                # None！
print("Original Result：")
print(L)                # L 自己被排好了
````

## 4. 用 `lambda` 排序 tuple/list


````python
s = [('a', 9), ('b', 8), ('c', 7)]

t = sorted(s, key=lambda v: v[1])                 # 依第 2 欄
print(t)
t = sorted(s, key=lambda v: v[1], reverse=True)   # 依第 2 欄由大到小
print(t)
u = sorted(s, key=lambda w: w[0])                 # 依第 1 欄
print(u)
````

## 5. 雙重排序

「分數由大到小，分數相同時名字由小到大」：
利用 tuple 比較的特性，數值欄位加負號。


````python
students = [("Bob", 80), ("Amy", 90), ("Eve", 80), ("Dan", 90)]

result = sorted(students, key=lambda x: (-x[1], x[0]))
for name, score in result:
    print(name, score)
````

## 6. 自定義比較函數：`functools.cmp_to_key`

當排序規則太複雜、不容易寫成 key 時，
可以寫傳統的比較函數（回傳 -1, 0, 1），再用 `cmp_to_key` 轉換。


````python
import functools

def cmp(x, y):
    if x[1] < y[1]:
        return -1
    elif x[1] > y[1]:
        return 1
    else:
        return 0

s = [[1, 2], [3, 4], [2, 3]]
s.sort(key=functools.cmp_to_key(cmp))
print(s)
````

````python
# 複雜範例：先比總和，總和相同比第一個元素「由大到小」
import functools

def cmp2(x, y):
    if sum(x) != sum(y):
        return sum(x) - sum(y)     # 正數表示 x 排後面
    return y[0] - x[0]

s = [[1, 4], [3, 2], [2, 3], [5, 0]]
s.sort(key=functools.cmp_to_key(cmp2))
print(s)
````

## 7. 練習題

給定資料 `[("apple", 3, 20), ("banana", 3, 15), ("cherry", 5, 15)]`
（名稱、等級、價格），請依「等級由大到小、價格由小到大、名稱由小到大」排序。

期望輸出：

```text
('cherry', 5, 15)
('banana', 3, 15)
('apple', 3, 20)
```


````python
# 練習作答區
items = [("apple", 3, 20), ("banana", 3, 15), ("cherry", 5, 15)]
````

### 參考答案

````python
items = [("apple", 3, 20), ("banana", 3, 15), ("cherry", 5, 15)]
for it in sorted(items, key=lambda x: (-x[1], x[2], x[0])):
    print(it)
````
