# Python 程式撰寫原則

本教材整理幾個適合程式設計競賽與初學者練習的 Python 撰寫原則：

1. 盡量使用 Python 內建函式與標準函式庫
2. 維持一致的程式設計風格與變數命名
3. 重複出現三次以上的程式碼，建議定義為函式
4. 能用表格，就不要寫大量 `if`
5. 常數放在程式最前面
6. 選擇清楚的命名方式，避免容易混淆的字元
7. 至少連續測試兩筆資料，避免變數沒有重設


## 1. 盡量使用 Python 內建函式庫

Python 已經提供很多好用的內建函式與標準函式庫。

在比賽中，善用這些工具可以減少程式碼長度、降低寫錯的機率、提高解題速度，讓程式更容易閱讀。


### 不建議：自己手寫 permutation

下面這段程式可以產生排列，但程式較長，也比較容易寫錯。

````python
def permutation(lst):
    if len(lst) == 0:
        return []

    if len(lst) == 1:
        return [lst]

    result = []

    for I in range(len(lst)):
        m = lst[I]
        remain = lst[:I] + lst[I + 1:]

        for p in permutation(remain):
            result.append([m] + p)

    return result


data = [1, 2, 3]
result = permutation(data)

print(result)
````

### 建議：使用 `itertools.permutations`

Python 標準函式庫已經提供 `permutations`，通常比較簡潔、穩定，也比較不容易寫錯。

````python
from itertools import permutations

data = [1, 2, 3]

result = list(permutations(data))

print(result)
````

### 常見好用的 Python 內建工具

| 功能 | 寫法 |
|---|---|
| 排序 | `sorted(data)` |
| 最大值 | `max(data)` |

| 去除重複 | `set(data)` |
| 排列 | `itertools.permutations(data)` |

## 2. 能用表格，就不要寫大量 `if`

如果題目只是做「代碼對應資料」，例如月份數字對應月份名稱、分數對應等級、星期數字對應星期名稱，通常可以用 `list` 或 `dict` 取代大量 `if`。


### 不建議：大量 if

````python
def get_month_name(month):
    if month == 1:
        return "January"
    if month == 2:
        return "February"
    if month == 3:
        return "March"
    if month == 4:
        return "April"
    if month == 5:
        return "May"
    if month == 6:
        return "June"
    if month == 7:
        return "July"
    if month == 8:
        return "August"
    if month == 9:
        return "September"
    if month == 10:
        return "October"
    if month == 11:
        return "November"
    if month == 12:
        return "December"


print(get_month_name(2))
````

### 建議：使用 list 表格

````python
MONTH_NAME = [
    "None",
    "January",
    "February",
    "March",
    "April",
    "May",
    "June",
    "July",
    "August",
    "September",
    "October",
    "November",
    "December"
]

month = 2

print(MONTH_NAME[month])
````

### 另一種寫法：使用 dictionary

如果代碼不是連續數字，或資料對應比較複雜，可以使用 `dict`。

````python
MONTH_NAME = {
    1: "January",
    2: "February",
    3: "March",
    4: "April",
    5: "May",
    6: "June",
    7: "July",
    8: "August",
    9: "September",
    10: "October",
    11: "November",
    12: "December"
}

month = 12

print(MONTH_NAME[month])
````

## 3. 不要使用太複雜的一行寫法

程式不是越短越好。  
在比賽中，程式應該要讓自己容易檢查、容易除錯。  
如果一行公式太長，可以拆成多行。


### 不建議：一行塞太多邏輯

````python
def func(t):
    return t * 2


x = 3
y = 4
t = 5
s = 0.2

result = (x * y) + func(t) / (1 - s)

print(result)
````

### 建議：拆成中間變數

拆成多行之後，每一行的意義比較清楚，也比較方便印出來檢查。

````python
def func(t):
    return t * 2


x = 3
y = 4
t = 5
s = 0.2

base = x * y
temp = func(t)
result = base + temp / (1 - s)

print(result)
````

## 4. 重複出現三次以上的程式，建議定義成函式

如果同樣的邏輯一直重複出現，代表它應該被整理成函式。好處是減少重複程式碼、降低改錯時漏改的機率，並讓主程式更乾淨。


### 不建議：重複寫相同邏輯

````python
score1 = 85
score2 = 72
score3 = 59

if score1 >= 60:
    print("Pass")
else:
    print("Fail")

if score2 >= 60:
    print("Pass")
else:
    print("Fail")

if score3 >= 60:
    print("Pass")
else:
    print("Fail")
````

### 建議：定義函式

````python
def print_result(score):
    if score >= 60:
        print("Pass")
    else:
        print("Fail")


score1 = 85
score2 = 72
score3 = 59

print_result(score1)
print_result(score2)
print_result(score3)
````

## 5. 常數放在程式最前面

如果程式中有固定數值，例如及格分數、最大資料量、方向陣列、月份名稱、商品價格表，建議放在程式最前面，並使用大寫命名。


### 不建議：數字直接寫在程式中

這種寫法稱為 magic number。當規則改變時，可能要到處找數字。

````python
score = 87

if score >= 60:
    print("Pass")
else:
    print("Fail")
````

### 建議：先定義常數

````python
PASS_SCORE = 60

score = 87

if score >= PASS_SCORE:
    print("Pass")
else:
    print("Fail")
````

## 6. 至少連續測試相同的兩筆資料

很多錯誤只測一筆資料看不出來，特別是變數沒有歸零、list 沒有清空、set 沒有重設，或暫存資料留到下一筆測資。


### 題目

請輸出兩次 `1 + 2 + 3 + ... + 10` 的值。

正確輸出應該是：

```text
55
55
```


### 錯誤寫法：變數沒有重設

`s` 放在外面，所以第二次會接續第一次的結果。

````python
s = 0

for _ in range(2):
    for I in range(1, 11):
        s = s + I

    print(s)
````

### 正確寫法：每一筆資料開始前重新設定變數

`s = 0` 應該放在每一筆資料的迴圈裡面。

````python
for _ in range(2):
    s = 0

    for I in range(1, 11):
        s = s + I

    print(s)
````

## 7. 選擇清楚的命名方式

初學者常會混淆英文大寫 `O` 與數字 `0`、小寫 `l` 與數字 `1`、大寫 `I` 與小寫 `l`、符號 `|` 與大寫 `I`。因此，變數名稱要盡量清楚。


### 不建議：使用容易混淆的變數名稱

下面這些名稱不一定語法錯，但容易看錯。

````python
l = 10
I = 20
O = 30
number0 = 40

print(l, I, O, number0)
````

### 建議：使用有意義的名稱

例如處理二維表格時，用 `row`、`col` 會比 `i`、`j` 更容易理解。

````python
R = 3
C = 4

for row in range(R):
    for col in range(C):
        print(row, col)
````

## 8. 維持一致的程式風格

同一份程式應該維持一致的命名與排版。

建議：

- 變數名稱使用小寫與底線，例如 `student_score`
- 常數使用大寫，例如 `PASS_SCORE`
- 函式名稱使用小寫與底線，例如 `get_month_name`
- 縮排固定使用 4 個空白
- 不要一份程式中混用不同命名風格


````python
PASS_SCORE = 60


def get_result(score):
    if score >= PASS_SCORE:
        return "Pass"
    else:
        return "Fail"


student_score = 85
result = get_result(student_score)

print(result)
````
